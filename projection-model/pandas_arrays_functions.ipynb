```python
# If we want to add a year to every players age we can create a function to do so
```

```python
def add_year(current_age):
    new_age = current_age + 1
    return new_age
```

```python
df.info()
```

```python
# Use apply method to apply functions
df['age'].apply(add_year)
```

```python
def ppg_calc(row):
    gp = row[3]
    pts = row[17]
    ppg = pts/gp
    return ppg
```

```python
df.apply(ppg_calc, axis = 1)
```

```python
def ft_calc(row):
    ft_p = (row['ftm']/ (row['fta'] + 0.00001))
    return round(ft_p,1)
```

```python
df['ft_percent'] = df.apply(ft_calc, axis=1)
```

```python
df.sample(10)
```

```python
# To find career ft %
```

```python
df.groupby('player_name').sum().apply(ft_calc, axis=1)
```
