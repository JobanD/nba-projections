```python
# Merging and Joining Dataframes
```

```python
df1 = pd.read_csv('../nba-stats-csv/player_general_traditional_per_game_data.csv')
df2 = pd.read_csv('../nba-stats-csv/player_id_player_name.csv')
```

```python
df1.head(5)
```

```python
df2.tail(5)
```

```python
# We can see df1 and df2 have different columns
# Using the common col of player_id we can merge these dfs!!!
```

```python
merged_df = pd.merge(df1, df2, on='player_id')
```

```python
merged_df.sample(10)
```

```python
# to specify certain cols in the merge specify the cols in the merge method parameters
```

```python
merged_df2 = pd.merge(df2, df1[['player_id', 'season_id', 'ast', 'stl', 'pts']], on='player_id')
merged_df2.sample(10)
```

```python
# Can specify right, left, outer merge types using how parameter built into merge method
```
